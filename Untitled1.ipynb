```python
import pandas as pd
df=pd.read_csv("covid19_tweets.csv")
df.head()
```

```python
import re
def clean(s):
    s=s.replace(r'<lb>',"\n")
    s=re.sub(r'<br */*>', "\n", s)
    s=s.replace("&lt;","<").replace("&gt;", ">").replace("&amp;","&")
    s=s.replace("&amp;","&")
    s=re.sub(r'\(*https*://[^\)]*\)*',"", s)
    s=re.sub(r'_+',' ',s)
    s=re.sub(r'"+','"',s)
    return str(s)
```

```python
df["text_clean"]=''
for i, row in df.iterrows():
    df.at[i,"text_clean"]= clean(row.text)
```
